```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
```

```python
dataframe = pd.read_csv('passengers.csv',usecols=[1], engine='python', skipfooter=3)
```

```python
dataframe.tail(10)
```

```python
dataset = dataframe.values
```

```python
dataframe.values
```

```python
dataset = dataset.astype('float32')
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
dataset = scaler.fit_transform(dataset)
```

```python
dataset
```

```python
train_size = int(len(dataset) * 0.70)
```

```python
test_size = len(dataset) - train_size 
```

```python
test_size
```

```python
train = dataset[0:train_size,:]
```

```python
train
```

```python
test = dataset[train_size:len(dataset),:]
```

```python
def create_dataset(dataset, look_back=1):
    dataX, dataY = [], []
    for i in range(len(dataset)-look_back-1):
        a = dataset[i:(i+look_back), 0]
        dataX.append(a)
        dataY.append(dataset[i + look_back, 0])
    return np.array(dataX), np.array(dataY)
```

```python
look_back = 1
trainX, trainY = create_dataset(train, look_back)
testY, textY = create_dataset(test, look_back)
```

```python
print(testY)
```

```python
model = Sequential()
```

```python
model.add(LSTM(16, input_shape=(None, look_back)))
model.add(Dense(1))
model.compile(loss='mse', optimizer='rmsprop')
```

```python
model.fit(trainX, trainY, epochs=50)
```

```python
trainPredict = model.predict(trainX)
testPredict = model.predict(textX)
```
